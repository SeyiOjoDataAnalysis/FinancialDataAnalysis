The time value of money concept refers to the potential of money today to be invested, earn interest, and be worth more in the future. 

rate = rate of return
nper = timespan of investment
pmt = fixed payment at beginning/end of each period.
fv = future value of investment

```python
import numpy_financial as npf

present_value = npf.pv(rate=0.03, nper=5, pmt=0, fv=100)
present_value
```

The negative sign indicates that 100 units would need to invested (i.e. deducted) to meet future obligations. 

```python
future_value = npf.fv(rate = 0.03, nper = 5, pmt = 0, pv = -100)
future_value
```

```python
import numpy_financial as npf
import matplotlib.pyplot as plt


interest_rate = 0.03
nper = 5
pmt = 0
initial_investment_range = range(1, 101) 
future_values = []

# Calculation of future values 
for initial_investment in initial_investment_range:
    future_value = npf.fv(rate=interest_rate, nper=nper, pmt=pmt, pv=-initial_investment)
    future_values.append(future_value)


plt.plot(initial_investment_range, future_values, label='Future Values')
plt.xlabel('Initial Investment')
plt.ylabel('Future Value')
plt.title('Future Value vs. Initial Investment')
plt.grid(True)
plt.show()
```

```python
def get_future_value(rate, nper, pmt, pv):
    future_value = npf.fv(rate = rate, nper = nper, pmt = 0, pv = pv)    
    print(f'If you invested {(pv**2)**0.5} pounds, at a {rate}% interest rate, in {nper} years, the value would increase to {future_value}')
    
    


def get_present_value(rate, nper, pmt, fv):
    present_value = npf.pv(rate = rate, nper = nper, pmt = 0, fv = fv)
    
    
    return present_value




```
